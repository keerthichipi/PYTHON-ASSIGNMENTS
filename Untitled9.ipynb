```python
word = "pyTHoN"
result = ""

for ch in word:
    if ch.islower():
        result += ch.upper()
    else:
        result += ch.lower()

print(result)
```

```python
# Take input from the user
word = input("Enter a word: ")

# Variable to store the longest substring without repetition
longest = ""

# Loop through each character index in the word
for i in range(len(word)):

    # Temporary string to store current substring
    temp = ""

    # Loop from current index to the end of the word
    for j in range(i, len(word)):

        # If character is not already present in temp
        if word[j] not in temp:

            # Add character to temp
            temp = temp + word[j]

        else:
            # Break the loop if repetition occurs
            break

    # If temp length is greater than longest, update longest
    if len(temp) > len(longest):
        longest = temp

# Print the longest substring without repetition
print("Longest word without repetition:", longest)
```
