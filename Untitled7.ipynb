```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
num = int(input("Enter a number: "))

prev = num - 1

while prev > 1:
    if is_prime(prev):
        print("Previous Prime Number:", prev)
        break
    prev -= 1
else:
    print("No previous prime number found")
```

```python
num = int(input("Enter a number: "))

# Previous prime
prev = num - 1
while prev > 1 and not is_prime(prev):
    prev -= 1

# Next prime
next_p = num + 1
while not is_prime(next_p):
    next_p += 1

print("Previous Prime:", prev if prev > 1 else "Not available")
print("Next Prime:", next_p)

# Nearest prime
if prev > 1:
    if (num - prev) < (next_p - num):
        nearest = prev
    elif (num - prev) > (next_p - num):
        nearest = next_p
    else:
        nearest = f"{prev} and {next_p} (both are equally near)"
else:
    nearest = next_p

print("Nearest Prime:", nearest)
```
