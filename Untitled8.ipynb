```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


def nearest_prime(num):
    if num <= 2:
        return 2

    left = num - 1
    right = num + 1

    while True:
        if is_prime(left):
            return left
        if is_prime(right):
            return right
        left -= 1
        right += 1


# Main program
num = int(input("Enter a number: "))
print("Nearest Prime Number:", nearest_prime(num))
```
